```python
import numpy as np
import scipy as sp
import scipy.stats as sts
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 200)
```

```python
alpha = 1
beta = 2
distribution = sts.gamma(a=alpha, scale=1/beta)
plt.figure(figsize=(10, 6))
plt.plot(x, distribution.pdf(x), label='a=1, b=2')

alpha = 2
beta = 2
distribution = sts.gamma(a=alpha, scale=1/beta)
plt.plot(x, distribution.pdf(x), label='a=2, b=2')

alpha = 3
beta = 1.5
distribution = sts.gamma(a=alpha, scale=1/beta)
plt.plot(x, distribution.pdf(x), label='a=1, b=1')

plt.legend()
plt.show()
```

```python
samples = distribution.rvs(size=100)

plt.figure(figsize=(10, 6))
plt.hist(samples, density=True)
plt.plot(x, distribution.pdf(x))

plt.show()
```
